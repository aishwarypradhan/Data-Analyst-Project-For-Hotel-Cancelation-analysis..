# We Have To See Here in Which Hotel there are Most No Of Canceled has Happended

```python
plt.figure(figsize = (8,4))
ax1= sns.countplot(x = "hotel", hue = "is_canceled", data = df, palette = "Blues")
legend_labels,_ = ax1. get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title("Reservation status in different hotels",size = 20)
plt.xlabel("hotel")
plt.ylabel("number of reservations")
plt.show()
```

```python
##in Resort Hotel We Will See The Cancel  Not Cancel In Percentage.........................................
resort_hotel=df[df["hotel"]=="Resort Hotel"]
resort_hotel["is_canceled"].value_counts(normalize=True)
```

```python
##in City Hotel We Will See The Cancel  Not Cancel In Percentage.........................................
city_hotel=df[df["hotel"]=="City Hotel"]
city_hotel["is_canceled"].value_counts(normalize=True)
```

```python
resort_hotel=resort_hotel.groupby("reservation_status_date")[["adr"]].mean()
city_hotel=city_hotel.groupby("reservation_status_date")[["adr"]].mean()
```

```python
#######Here We Will Create Average Daily Rate In City And Resort Hotel..................................

plt.figure(figsize=(22,8))
plt.title("Average Daily Rate in City and Resort Hotel",fontsize=30,color="green")
plt.plot(resort_hotel.index,resort_hotel["adr"],label="Resort Hotel") 
plt.plot(city_hotel.index,city_hotel["adr"],label="City Hotel")
plt.legend(fontsize=20)
plt.show()
```

```python
df["month"]=df["reservation_status_date"].dt.month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x="month",hue="is_canceled",data=df,palette="bright")
legend_labels,_ = ax1. get_legend_handles_labels()
ax1.legend(bbox_to_anchor=(1,1))
plt.title( 'Reservation status per month', size = 20)
plt.xlabel( "month")
plt.ylabel( "number of reservations")
plt.legend(['not canceled', 'canceled'])
plt.show()

```

```python
# Here We Will See Which Month The Price Was Low In This Case We Will Find Out Why August Month will High Reservation
#because OF August Month reservation hotel  Rate in low so the is_Not_Canceled Is High 


plt.figure(figsize=(16,8))
plt.title( 'ADR per month', size = 20)
sns.barplot("month","adr",data=df[df["is_canceled"]==1].groupby("month")[["adr"]].sum().reset_index())
plt.show()
```

Here Its proved That When High Price Than Our Cancelation Is high Here We Can See In january Month hotel Rate Is 
High So the Cancelation Is High people Are Cancelling when The Price Is Going up but when the Price is low,We Can See 
in August month price is low and cancelation is low when Low Price peoples Are Not Canceling Their Reservation....
